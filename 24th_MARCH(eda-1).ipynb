#### `Q2`. How did you handle missing data in the wine quality data set during the feature engineering process? Discuss the advantages and disadvantages of different imputation techniques.


 * Missing data is a common problem in datasets, and it can be caused by various factors such as data entry errors, data loss during transfer, or incomplete surveys. There are several ways to handle missing data, including:

1. Deletion: This involves removing rows or columns with missing data. The advantage of this method is that it is straightforward and does not require any assumptions about the missing data. However, it can lead to a loss of information and potentially biased results if the missing data is not random.

2. Mean/median imputation: This involves replacing missing values with the mean or median of the observed values for that feature. The advantage of this method is that it is simple and can work well if the missing data is not too extensive. However, it can underestimate the variance of the feature and can introduce bias in the data if the missing data is not missing at random.

3. Mode imputation: This involves replacing missing categorical data with the mode or most frequently occurring value of that feature. The advantage of this method is that it is simple and can work well for categorical data. However, it can be biased if the missing data is not missing at random.

###`Q5`. Load the wine quality data set and perform exploratory data analysis (EDA) to identify the distribution of each feature. Which feature(s) exhibit non-normality, and what transformations could be applied to these features to improve normality?



* To perform EDA on the wine quality dataset and identify the distribution of each feature, we can start by loading the dataset into a tool like Python or R and visualizing the distributions using histograms, density plots, or Q-Q plots.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

wine_df = pd.read_csv('https://raw.githubusercontent.com/aniruddhachoudhury/Red-Wine-Quality/master/winequality-red.csv')
wine_df.head()
```

```python
wine_df.describe()
```

```python
wine_df.hist(figsize=(12, 12), bins=20 )
plt.show()
```

```python
# Plot density plots of all variables
wine_df.plot(kind='density', subplots=True, layout=(4, 3), sharex=False, figsize=(12, 12))
plt.show()
```

```python
from scipy import stats
for col in wine_df.columns:
    plt.figure()
    stats.probplot(wine_df[col], dist="norm", plot=plt)
    plt.title(col)
    plt.show()
```

* From the histograms and density plots, we can see that some of the features exhibit non-normality. For example, "volatile acidity", "citric acid", "residual sugar", "chlorides", "free sulfur dioxide", and "total sulfur dioxide" all have distributions that are skewed to the right (i.e., have long tails on the positive side).

* To improve normality in these features, we could apply a transformation such as log or square root. For example, we could transform "volatile acidity" using a log transformation as follows:

```python
# Apply log transformation to volatile acidity
wine_df['volatile acidity'] = np.log(wine_df['volatile acidity'])
```

```python

# Plot histogram of transformed variable
plt.hist(wine_df['volatile acidity'], bins=20)
plt.show()
```

```python
# Plot Q-Q plot of transformed variable
stats.probplot(wine_df['volatile acidity'], dist="norm", plot=plt)
plt.title('volatile acidity (log transformed)')
plt.show()
```

###`Q6`. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of features. What is the minimum number of principal components required to explain 90% of the variance in the data?

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
wine_scaled = scaler.fit_transform(wine_df)
```

```python
pca = PCA()
pca.fit(wine_scaled)
```

```python
# Extract the eigenvalues and explained variance ratios
eigenvalues = pca.explained_variance_
explained_variances = pca.explained_variance_ratio_
```

```python
plt.plot(range(1, len(explained_variances)+1), explained_variances, 'ro-', linewidth=2)
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Proportion of Variance Explained')
plt.show()
```

```python
# Calculate cumulative variance explained
cumulative_variances = np.cumsum(explained_variances)

# Find the number of components required to explain 90% of the variance
n_components = np.argmax(cumulative_variances >= 0.9) + 1

print("Number of principal components required to explain 90% of the variance: ", n_components)
```
